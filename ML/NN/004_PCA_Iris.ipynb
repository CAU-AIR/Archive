### Principal Component Analysis(PCA)



```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

### PCA 3D

```python
iris = load_iris()

pca = PCA(n_components=3)
features = pca.fit(iris.data).transform(iris.data)

fig = plt.figure()
ax = Axes3D(fig)

x = features[:, 0]
y = features[:, 1]
z = features[:, 2]

img = ax.scatter(x, y, z, c=iris.target, cmap=plt.cm.rainbow)
plt.legend(img.legend_elements()[0], iris.target_names)
plt.show()
```

### PCA 2D

```python
pca = PCA(n_components=2)
features = pca.fit(iris.data).transform(iris.data)

x = features[:, 0]
y = features[:, 1]

plt.figure()
img = plt.scatter(x, y, c=iris.target, cmap=plt.cm.rainbow)
plt.legend(img.legend_elements()[0], iris.target_names)
plt.show()
```
