## Experimental results by train & test size using Dimension Reduction

학습 데이터 및 테스트 사이즈에 따른 실험 결과

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
```

## LDA Feature 2 & QDA Classification

```python
test_acc = []
train_time = []

features, target = make_classification(n_samples=15000, n_features=4, n_informative=4, n_redundant=0, n_classes=3, n_clusters_per_class=1, random_state=0)

lda = LinearDiscriminantAnalysis(n_components=2)
features = lda.fit_transform(features, target)
target_names = ["0", "1", "2"]
```

### Test Size 10%

```python
max_seed = 10

avg_test_acc = []
avg_time = []

predict_list = []
label_list = []

for seed in range(max_seed):

    x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=seed, shuffle=True)

    qda = QuadraticDiscriminantAnalysis()
    qda.fit(x_train, y_train)

    predict = qda.predict(x_test)

    predict_list.extend(predict)
    label_list.extend(y_test)

    start = time.perf_counter()
    test_score = qda.score(x_test, y_test)
    end = time.perf_counter() - start

    avg_test_acc.append(test_score)
    avg_time.append(end)

print("Train size : ", len(y_train), " / Test size : ", len(y_test))
test_avg = np.array(avg_test_acc).mean()
time_avg = np.array(avg_time).mean()

test_acc.append(test_avg)
train_time.append(time_avg)

cm = confusion_matrix(label_list, predict_list, labels=qda.classes_, normalize="true")
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)

cm = confusion_matrix(label_list, predict_list, labels=qda.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
```
