## Experimental results by train & test size

학습 데이터 및 테스트 데이터 사이즈에 따른 실험 결과

<span style="color: #2D3748; background-color:#fff5b1;">테스트 사이즈가 점점 증가하고 학습 데이터가 점점 감소하는 경우 정확도는 감소한다. 하지만 분류 시간의 경우 학습 데이터와 테스트 데이터의 비율에 따라 증가하다가 감소한다.</span>

<span style="color: #519800">ex) 전체 데이터의 개수가 100개인 경우 train 80개, test 20개의 경우 1,600번의 거리 계산이 수행된다. Train 50개, test 50개의 경우 2,500번의 거리 계산이 수행된다. 하지만 train 10개, test 90개의 경우 900번의 거리 계산이 수행되기 때문에 분류 시간이 증가하다가 감소하게 된다. </span>

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
test_acc = []
train_time = []

features, target = make_classification(n_samples=15000, n_features=4, n_informative=4, n_redundant=0, n_classes=3, n_clusters_per_class=1, random_state=0)
target_names = ["0", "1", "2"]
```

### Test Size 10%

```python
max_seed = 10

avg_test_acc = []
avg_time = []

predict_list = []
label_list = []

for seed in range(max_seed):

    x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=seed, shuffle=True)

    lda = LinearDiscriminantAnalysis()
    lda.fit(x_train, y_train)

    predict = lda.predict(x_test)

    predict_list.extend(predict)
    label_list.extend(y_test)

    start = time.perf_counter()
    test_score = lda.score(x_test, y_test)
    end = time.perf_counter() - start

    avg_test_acc.append(test_score)
    avg_time.append(end)

print("Train size : ", len(y_train), " / Test size : ", len(y_test))
test_avg = np.array(avg_test_acc).mean()
time_avg = np.array(avg_time).mean()

test_acc.append(test_avg)
train_time.append(time_avg)

cm = confusion_matrix(label_list, predict_list, labels=lda.classes_, normalize="true")
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)

cm = confusion_matrix(label_list, predict_list, labels=lda.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
```
