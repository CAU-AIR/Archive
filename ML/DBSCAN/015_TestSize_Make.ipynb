## Experimental results by train & test size

학습 데이터 및 테스트 데이터 사이즈에 따른 실험 결과

<span style="color: #2D3748; background-color:#fff5b1;">테스트 사이즈가 점점 증가하고 학습 데이터가 점점 감소하는 경우 정확도는 감소한다. 하지만 분류 시간의 경우 학습 데이터와 테스트 데이터의 비율에 따라 증가하다가 감소한다.</span>

<span style="color: #519800">ex) 전체 데이터의 개수가 100개인 경우 train 80개, test 20개의 경우 1,600번의 거리 계산이 수행된다. Train 50개, test 50개의 경우 2,500번의 거리 계산이 수행된다. 하지만 train 10개, test 90개의 경우 900번의 거리 계산이 수행되기 때문에 분류 시간이 증가하다가 감소하게 된다. </span>

```python
import time
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
total_acc = []
total_time = []

features, target = make_classification(n_samples=15000, n_features=4, n_informative=4, n_redundant=0, n_classes=3, n_clusters_per_class=1, random_state=0)
scaler = MinMaxScaler()
features = scaler.fit_transform(features)
target_names = ["0", "1", "2", "outlier"]
```

### Dataset Size 10%

```python
epsilon = 0.1
point = 10

x_train, _, y_train, _ = train_test_split(features, target, train_size=0.1, random_state=0, shuffle=True)

db = DBSCAN(eps=epsilon, min_samples=point, n_jobs=-1)

start = time.perf_counter()
db.fit(x_train, y_train)
end = time.perf_counter() - start

predict = db.labels_
predict[db.labels_ == -1] = 3

acc = metrics.accuracy_score(y_train, db.labels_)

print("Train size : ", len(y_train))
total_acc.append(acc)
total_time.append(end)

cm = confusion_matrix(y_train, predict, normalize="true")
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= target_names)
disp.plot(cmap=plt.cm.Blues)

cm = confusion_matrix(y_train, predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= target_names)
disp.plot(cmap=plt.cm.Blues)
```
