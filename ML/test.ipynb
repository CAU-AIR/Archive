```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_stack = pd.DataFrame(dict(
    score = [86.55, 86.19, 81.89, 80.29, 80.27]
))

sns.set(style="white", rc={"lines.linewidth": 3})

fig, ax1 = plt.subplots(figsize=(15, 10))

sns.barplot(x=["LDA-ResNet-18", "1-NN(C)-ResNet-18", "1-NN(E)-ResNet-18", "NCM-ResNet-18", "LDA-ResNet-18"],
            y="score",
            data=df_stack,
            label="Accuracy",
            color='#004488',
            ax=ax1)

ax1.set_ylabel("Accuracy")

lines_1, labels_1 = ax1.get_legend_handles_labels()


width_scale = 1

for bar in ax1.containers[0]:
    bar.set_width(bar.get_width() * width_scale)


plt.legend(lines_1, labels_1, loc='center right', bbox_to_anchor=(1.25, 0.5))
plt.show()
sns.set()
```
