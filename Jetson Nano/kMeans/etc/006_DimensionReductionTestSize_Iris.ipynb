## Experimental results by train & test size using Dimension Reduction

학습 데이터 및 테스트 사이즈에 따른 실험 결과

<span style="color: #2D3748; background-color:#fff5b1;">테스트 사이즈가 점점 증가하고 학습 데이터가 점점 감소하는 경우 정확도는 감소한다. 하지만 분류 시간의 경우 학습 데이터와 테스트 데이터의 비율에 따라 증가하다가 감소한다. PCA를 적용하지 않은 방법과 유사한 형태의 실험 결과를 보여준다.</span>

```python
import time
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## k-Means with PCA feature 3

```python
total_acc = []
total_time = []

iris = load_iris()

pca = PCA(n_components=3)
features = pca.fit(iris.data).transform(iris.data)
target = iris.target
target_names = iris.target_names
```

### Dataset Size 10%

```python
max_seed = 10

avg_acc = []
avg_time = []

predict_list = []
label_list = []

for seed in range(max_seed):

    x_train, _, y_train, _ = train_test_split(features, target, train_size=0.1, random_state=seed, shuffle=True)

    k = 3

    k_means = KMeans(n_clusters=k, random_state=seed)
    start = time.perf_counter()
    k_means.fit(x_train, y_train)
    end = time.perf_counter() - start

    predict = k_means.labels_
    score = metrics.accuracy_score(y_train, k_means.labels_)

    predict_list.extend(predict)
    label_list.extend(y_train)

    avg_acc.append(score)
    avg_time.append(end)

print("Train size : ", len(y_train))
acc_avg = np.array(avg_acc).mean()
time_avg = np.array(avg_time).mean()

total_acc.append(acc_avg)
total_time.append(time_avg)

cm = confusion_matrix(label_list, predict_list, normalize="true")
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= target_names)
disp.plot(cmap=plt.cm.Blues)

cm = confusion_matrix(label_list, predict_list)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= target_names)
disp.plot(cmap=plt.cm.Blues)
```
